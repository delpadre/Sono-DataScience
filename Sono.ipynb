```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

caminho_arquivo = ('/content/CheckPoint/Sleep_Efficiency_Updated.csv')
df = pd.read_csv(caminho_arquivo)


numerical_df = df.select_dtypes(include=['float64', 'int64'])

correlation_matrix = numerical_df.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Matriz de Correlação')
plt.show()
```

# **Explicação do Código**

Esse código realiza uma análise de correlação entre diferentes variáveis numéricas. A correlação mede a relação entre duas variáveis: se elas aumentam ou diminuem juntas, ou se uma aumenta enquanto a outra diminui. Vamos dividir em etapas simples:

**Leitura do arquivo CSV**: O código começa lendo um arquivo de dados.
Criando um DataFrame com as informações.

**Verificação dos tipos de dados**: O código verifica que tipo de dado está presente em cada coluna (por exemplo, números inteiros, números decimais, texto, etc.).

**Seleção de colunas numéricas**: Apenas as colunas que contêm números (do tipo float64 ou int64) são selecionadas. Isso é feito porque a correlação só faz sentido entre números.

**Cálculo da matriz de correlação**: A matriz de correlação é uma tabela que mostra como cada variável numérica se relaciona com as outras. O valor da correlação varia de -1 a 1:

1 significa que duas variáveis aumentam ou diminuem juntas (correlação positiva perfeita).
0 significa que não há relação entre as variáveis.
-1 significa que quando uma variável aumenta, a outra diminui (correlação negativa perfeita).
Plotagem do gráfico: O código usa uma biblioteca de visualização para desenhar um gráfico de calor, onde as cores representam a força e a direção dessas correlações. Cores quentes (vermelho) indicam correlações fortes e positivas, enquanto cores frias (azul) indicam correlações negativas.

O boxplot é uma ferramenta visual que mostra a distribuição de um conjunto de dados, permitindo ver se há uma relação entre os dois fatores. Vamos explicar passo a passo o que isso significa e como interpretar o resultado:

**Caixa (Box)**: A caixa mostra os 50% centrais dos dados (entre o primeiro e o terceiro quartil, ou seja, do 25% ao 75% dos valores).

**Linha no meio (Mediana)**: A linha dentro da caixa indica a mediana (o valor central), que divide a metade inferior da metade superior dos dados.

**Extensões**: As linhas que saem da caixa mostram a extensão dos dados, mas até um certo limite.

**Outliers**: Pontos fora dos whiskers são valores atípicos, que são significativamente mais altos ou mais baixos do que o restante dos dados.

Outliers: Os pontos fora da caixa indicam pessoas com eficiências de sono incomuns (muito altas ou muito baixas) em comparação com a maioria dos outros.

```python
import pandas as pd
import matplotlib.pyplot as plt

caminho_arquivo = ('/content/CheckPoint/Sleep_Efficiency_Updated.csv')
df = pd.read_csv(caminho_arquivo)

media_duracao_sono = df['Sleep duration'].mean()
moda_duracao_sono = df['Sleep duration'].mode()[0]

print(f"Média da Duração do Sono: {media_duracao_sono:.2f} horas")
print(f"Duração de Sono mais frequente (moda): {moda_duracao_sono} horas")

plt.figure(figsize=(8, 6))

plt.bar(['Média', 'Moda'], [media_duracao_sono, moda_duracao_sono], color=['blue', 'orange'], width=0.5)

plt.title('Média e Moda da Duração do Sono', fontsize=16)
plt.xlabel('Métrica', fontsize=12)
plt.ylabel('Duração do Sono (horas)', fontsize=12)

for index, value in enumerate([media_duracao_sono, moda_duracao_sono]):
    plt.text(index, value + 0.1, f'{value:.2f}', ha='center', fontsize=12)

plt.tight_layout()
plt.show()
```

# **Explicação do Código**

Este gráfico de barras mostra dois valores importantes relacionados à duração do sono das pessoas no dataset:

Média da Duração do Sono: Representa o número médio de horas que as pessoas dormem. A "média" é calculada somando todas as durações de sono e dividindo pelo número total de pessoas. Isso nos dá uma ideia geral de quanto tempo, em média, as pessoas dormem. No gráfico, esse valor é mostrado pela barra azul.

Moda da Duração do Sono: Representa o valor que aparece mais vezes no conjunto de dados. Ou seja, é a quantidade de horas de sono mais comum entre todas as pessoas. A barra laranja no gráfico mostra esse valor.

As barras no gráfico têm rótulos no topo que mostram os números exatos da média e da moda, facilitando a visualização.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

caminho_arquivo = ('/content/CheckPoint/Sleep_Efficiency_Updated.csv')
df = pd.read_csv(caminho_arquivo)

# Verificando se as colunas necessárias existem
required_columns = ['REM sleep percentage', 'Light sleep percentage', 'Deep sleep percentage']
if all(column in df.columns for column in required_columns):
    # Criar um DataFrame com os percentuais de sono
    sleep_data = df[required_columns]

    # Criar um gráfico de dispersão relacionando a porcentagem de sono REM, sono leve e sono profundo
    plt.figure(figsize=(10, 6))

    # Gráfico de dispersão
    sns.scatterplot(data=sleep_data, x='Light sleep percentage', y='REM sleep percentage', label='Sono Leve vs REM', color='blue')
    sns.scatterplot(data=sleep_data, x='Deep sleep percentage', y='REM sleep percentage', label='Sono Pesado vs REM', color='orange')

    # Adicionando linhas de regressão
    sns.regplot(data=sleep_data, x='Light sleep percentage', y='REM sleep percentage', scatter=False, color='blue', ci=None, label='Regressão Leve')
    sns.regplot(data=sleep_data, x='Deep sleep percentage', y='REM sleep percentage', scatter=False, color='orange', ci=None, label='Regressão Pesada')

    # Configurações do gráfico
    plt.title('Relação entre Percentual de Sono REM, Sono Leve e Sono Pesado')
    plt.xlabel('Percentual de Sono Leve e Pesado (%)')
    plt.ylabel('Percentual de Sono REM (%)')
    plt.legend()
    plt.grid(True)
    plt.show()
else:
    print("Uma ou mais colunas necessárias não foram encontradas no dataset.")

```

**Eixo X (Horizontal)**: Representa o percentual de sono leve e profundo, medido em porcentagens (0% a 100%).
**Eixo Y (Vertical)**: Representa o percentual de sono REM, também em porcentagens.

**Pontos Azuis**: Correspondem à relação entre o percentual de sono leve e o percentual de sono REM. Cada ponto azul representa um conjunto de dados, indicando a quantidade de sono leve e a quantidade correspondente de sono REM de uma pessoa.

**Pontos Laranjas**: Correspondem à relação entre o percentual de sono pesado e o percentual de sono REM. Cada ponto laranja indica a quantidade de sono profundo e a quantidade correspondente de sono REM.

**Linha Azul (Sono Leve)**: Representa a tendência entre o percentual de sono leve e o sono REM. A inclinação da linha azul sugere que, conforme a porcentagem de sono leve aumenta, a porcentagem de sono REM tende a se manter estável, indicando que não há uma relação positiva clara.

**Linha Laranja (Sono Pesado)**: Mostra a tendência entre o percentual de sono profundo e o sono REM. A linha laranja tem uma inclinação negativa, indicando que, à medida que o sono profundo aumenta, o sono REM tende a diminuir. Isso pode indicar que um aumento no sono profundo está associado a uma redução no sono REM.

# **Quanto mais sono profundo eu tiver, menor sera meu sono rem?**

Sim, a relação entre o sono profundo e o sono REM (Movimento Rápido dos Olhos) é um aspecto interessante no sono. O sono é dividido em diferentes estágios, incluindo sono leve, sono profundo e sono REM.

Ciclos do Sono: O sono é organizado em ciclos de aproximadamente 90 minutos, que incluem estágios de sono leve, sono profundo e sono REM. Cada ciclo pode variar em duração e proporção dos diferentes estágios.

Sono Profundo: Este estágio é caracterizado por ondas cerebrais lentas e é essencial para a recuperação física e o restabelecimento do corpo. O sono profundo tende a ocorrer mais nas primeiras horas de sono.

Sono REM: Este estágio é importante para o processamento emocional, aprendizado e memorização. O sono REM geralmente ocorre mais nas últimas horas de sono.

Interação: Embora haja uma relação entre a quantidade de sono profundo e REM, não é uma regra rígida que a maior quantidade de sono profundo resulte em menos sono REM. Na verdade, muitas vezes, uma boa qualidade de sono permite uma proporção equilibrada de ambos os estágios.

Qualidade do Sono: Fatores como estresse, saúde física e hábitos de sono podem afetar tanto o sono profundo quanto o REM. Um sono de má qualidade pode levar a uma redução em ambos os estágios.

Portanto, enquanto um aumento no sono profundo pode potencialmente afetar a quantidade de sono REM, a interação entre os estágios de sono é complexa e pode variar de pessoa para pessoa. É importante focar na qualidade do sono como um todo, garantindo que você esteja passando por todos os estágios de maneira saudável.
